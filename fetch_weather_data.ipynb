```python
import pandas as pd
import requests
from datetime import datetime
```

```python
from datetime import datetime

latitude = 43.65107
longitude = -79.347015
hourly_vars = ["temperature_2m", "precipitation", "windspeed_10m", "cloudcover"]
timezone = "America%2FToronto"

# Today's date
today = datetime.today().strftime('%Y-%m-%d')

def fetch_weather_data_range(start_date, end_date):
    url = (
        f"https://archive-api.open-meteo.com/v1/archive?"
        f"latitude={latitude}&longitude={longitude}"
        f"&start_date={start_date}&end_date={end_date}"
        f"&hourly={','.join(hourly_vars)}"
        f"&timezone={timezone}"
    )
    print(f"Fetching from {start_date} to {end_date}...")
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data['hourly'])
        df['time'] = pd.to_datetime(df['time'])
        return df
    else:
        print(f"Failed: {response.status_code}")
        return pd.DataFrame()

# Yearly fetch from 2015 to 2024
dfs = [fetch_weather_data_range(f"{year}-05-01", f"{year}-12-31") for year in range(2015, 2025)]


# Fetch current year's data (2025)
dfs.append(fetch_weather_data_range("2025-01-01", today))

# Combine all
all_years = pd.concat(dfs, ignore_index=True)
```
